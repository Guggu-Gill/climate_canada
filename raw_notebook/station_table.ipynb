```python
import pandas as pd 
df_1=pd.read_csv("../data/climate-stations_1.csv")
df_2=pd.read_csv("../data/climate-stations_2.csv")
df_3=pd.read_csv("../data/climate-stations_3.csv")
df_4=pd.read_csv("../data/climate-stations_4.csv")
df_5=pd.read_csv("../data/climate-stations_5.csv")
```

```python
df=pd.concat((df_1,df_2,df_3,df_4,df_5),axis=0)
```

```python
accepted_columns=['x','y','STN_ID','STATION_NAME','PROV_STATE_TERR_CODE','ELEVATION','FIRST_DATE','LAST_DATE']
```

```python
df.dtypes
```

```python
df=df[accepted_columns]
```

```python
df
```

```python
df['FIRST_DATE'] = pd.to_datetime(df['FIRST_DATE']).dt.year
df['LAST_DATE'] = pd.to_datetime(df['LAST_DATE']).dt.year
```

```python
nu=df[df["PROV_STATE_TERR_CODE"]=="NU"]
pe=df[df["PROV_STATE_TERR_CODE"]=="PE"]
nt=df[df["PROV_STATE_TERR_CODE"]=="NT"]
nb=df[df["PROV_STATE_TERR_CODE"]=="NB"]
yt=df[df["PROV_STATE_TERR_CODE"]=="YT"]
```

```python
def three_large(nu,top_n=3):
    nu['age']=nu['LAST_DATE']-nu['FIRST_DATE']
    nu_3_lg=nu['age'].nlargest(top_n)
    nu = nu[nu['age'].isin(nu_3_lg)]
    return nu.drop('age',axis=1)
```

```python
nu=three_large(nu)
pe=three_large(pe)
nt=three_large(nt)
nb=three_large(nb)
yt=three_large(yt)
```
